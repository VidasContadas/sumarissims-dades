# Sumaríssims en grup

```python
import pandas as pd
import altair as alt


dades = pd.read_msgpack("data/processat.msg")

```

### Els 15 grups més grans 

(ordenats per nombre d'homes)

```python
dades.query("genere != '--'").\
      groupby(["ref_arxiu","genere"])["nom"].\
      count().\
      unstack("genere").\
      sort_values("Home",ascending=False).iloc[:15]
```

### El grup  45492

```python
grupo_45492 = dades.query("ref_arxiu == 45492")

grupo_45492
```

### Els encausats durant la republica

```python
grupo_1938 = dades.query("any_inicial == 1938")

grupo_1938
```

### Els 20 grups més grans pel nombre de dones

```python
dades.query("genere != '--'").\
                      groupby(["ref_arxiu","genere","any_inicial"])["nom"].count().\
                      unstack("genere").\
                      sort_values("Dona",ascending=False).iloc[:20]
```

### Les noms de dones mès comuns 

... i edat mitjana de les encausades fins a 1940. Com avui, els noms més populars canvien amb el temps. L'edat mitjana de les encausades amb nom Josefina és de 29 anys, amb nom Josefa 38.

```python
dades.query("genere == 'Dona' and edat <9999 and any_inicial< '1941-12-31'").\
      groupby(["nom"]).\
      aggregate({"genere" : "count", "edat": "mean"}).\
      rename(columns={'genere': 'nombre','edat': 'edat mitjana'}).\
      query("nombre> 20").\
      sort_values("edat mitjana",ascending=False).\
      style.format("{:0g}")
```
